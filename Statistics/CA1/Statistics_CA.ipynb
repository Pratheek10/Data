```r
#Model Building
#Initially we will build the model with all the input variables
model_all <- lm(deathRate ~.,data = modified_data)
par(mfrow = c(2,2))
summary(model_all)
plot(model_all)
```

```r
model_with_no_significance_removed <- lm(deathRate ~ incidenceRate+AvgHouseholdSize+povertyPercent
                     +PctHS18_24+PctBachDeg25_Over+PctEmpPrivCoverage+PctPrivateCoverage+PctWhite+PctOtherRace, data = modified_data)

par(mfrow = c(2,2))
summary(model_with_no_significance_removed)
plot(model_with_no_significance_removed)
```

```r
#Finding the influencial values in the dataset
library(olsrr)
ols_plot_cooksd_bar(model_with_no_significance_removed)
```

```r
model_with_no_significance_removed_removing_3040 <- lm(deathRate ~ incidenceRate+AvgHouseholdSize+povertyPercent
                     +PctHS18_24+PctBachDeg25_Over+PctEmpPrivCoverage+PctPrivateCoverage+PctWhite+PctOtherRace, data = modified_data[-3040,])

par(mfrow = c(2,2))
summary(model_with_no_significance_removed_removing_3040)
plot(model_with_no_significance_removed_removing_3040)
```

```r
#Even after removing all the unnecssary variabled and influecing data, accuracy has not increased
#As allmost all the variables are in percent,will try converting the other variables into percentage too
dummy_data <- modified_data
dummy_data$deathRate <- (dummy_data$deathRate/100000)*100
dummy_data$incidenceRate <- (dummy_data$incidenceRate/100000)*100
```

```r
#percentage variables
model_with_no_significance_removed_removing_3040_percenatage <- lm(deathRate ~ incidenceRate+AvgHouseholdSize+povertyPercent
                     +PctHS18_24+PctBachDeg25_Over+PctEmpPrivCoverage+PctPrivateCoverage+PctWhite+PctOtherRace, data = dummy_data[-3040,])

par(mfrow = c(2,2))
summary(model_with_no_significance_removed_removing_3040_percenatage)
plot(model_with_no_significance_removed_removing_3040_percenatage)
```

```r
#So even after converting them to percentage and removing the influcial variable, even though the residual error decreased, r squared value didnt increase
#So will try building the models on each socio economic factorS.
#Income : around 45 %
model_income <- lm(deathRate ~.,data = income)
par(mfrow = c(2,2))
summary(model_income)
plot(model_income)

#Education: around 47 %
model_education <- lm(deathRate ~.,data = education)
par(mfrow = c(2,2))
summary(model_education)
plot(model_education)

#Age : less than 26%
model_age <- lm(deathRate ~.,data = age)
par(mfrow = c(2,2))
summary(model_age)
plot(model_age)

#insurance: less than 45%
model_insurance <- lm(deathRate ~.,data = insurance)
par(mfrow = c(2,2))
summary(model_insurance)
plot(model_insurance)

#race: around 35%
model_race <- lm(deathRate ~.,data = race)
par(mfrow = c(2,2))
summary(model_race)
plot(model_race)

#household :30%
model_household <- lm(deathRate ~.,data = household)
par(mfrow = c(2,2))
summary(model_household)
plot(model_household)
```

```r
#When we built the model indidvisually with different socio factors, still the r square value was lesser than 50 for all,
#but it was 47 for education which was highest and 26 was the lowest for age.
#Will build a new dataframe with only significant values
new_dataset <- data.frame(deathRate,incidenceRate,AvgHouseholdSize,povertyPercent,PctHS18_24,PctBachDeg25_Over,PctEmpPrivCoverage,PctPrivateCoverage,PctWhite,PctOtherRace)
head(new_dataset)
```

```r
new_dataset <- new_dataset[-c(3040, 3045,337),]
model_all_new_dataset <- lm(deathRate ~.,data = new_dataset)
par(mfrow = c(2,2))
summary(model_all_new_dataset)
plot(model_all_new_dataset)
```

```r
#Will try transforming this if we can increase the R squqred value
#Have applied log for both input and output variables but still the r squared value has not increased
new_dataset <- new_dataset[-c(2620, 1514,2435),]
model_with_transformation_log <- lm(log(deathRate) ~ log(incidenceRate)+log(AvgHouseholdSize)+log(povertyPercent)+PctHS18_24+PctBachDeg25_Over+
                               log(PctEmpPrivCoverage)+log(PctPrivateCoverage)+PctWhite+PctOtherRace, data =new_dataset)

par(mfrow = c(2,2))
summary(model_with_transformation_log)
plot(model_with_transformation_log)
```

```r
#Using sqrt
model_with_transformation_sqrt <- lm(deathRate ~ sqrt(incidenceRate)+sqrt(AvgHouseholdSize)+sqrt(povertyPercent)+PctHS18_24+PctBachDeg25_Over+
                               sqrt(PctEmpPrivCoverage)+sqrt(PctPrivateCoverage)+PctWhite+PctOtherRace, data =new_dataset)

par(mfrow = c(2,2))
summary(model_with_transformation_sqrt)
plot(model_with_transformation_sqrt)
```

```r
#not only the below mentioned, but also tried with many different combinations but the r squared value is not increasing
new_dataset <- new_dataset[-c(3040, 3045,337),]
model_with_transformation_sqrt_log <- lm(sqrt(deathRate) ~ log(incidenceRate)+sqrt(AvgHouseholdSize)+log(povertyPercent)+PctHS18_24+PctBachDeg25_Over+
                               log(PctEmpPrivCoverage)+sqrt(PctPrivateCoverage)+PctWhite+PctOtherRace, data =new_dataset)

par(mfrow = c(2,2))
summary(model_with_transformation_sqrt_log)
plot(model_with_transformation_sqrt_log)
```
