```r
#preprocessing
#Removing unnecessary columns: As we dont require ID and no_of_patient
library(dplyr)
data <- data %>% select(-c(ID, No_Pation))
```

```r
#transforming the categorical variable into numerical
data$Gender <- ifelse(data$Gender == "M", 1, 0)
data$CLASS <- ifelse(data$CLASS == "Y", 1, 0)
```

```r
#checking the structure again
str(data)
```

```r
#just making sure that if any NA is present, that is removed
data <- na.omit(data)
```

```r
#Model building
install.packages('glmnet')
# Load the required library
library(glmnet)

#Splitting the data-set into train and test sets
set.seed(123)
train_index <- sample(nrow(data), nrow(data)*0.7)
train <- data[train_index, ]
test <- data[-train_index, ]
```

```r
#Fitting the logistic regression model on the training data with all the input variables
model_all <- glm(CLASS ~ ., data = train, family = binomial)
summary(model_all)
```

```r
par(mfrow=c(2,2))
plot(model_all)
```

```r
#this is for checking vif values as we need to remove values which have vif values more than 5-10
install.packages("car")
library(car)
print(vif(model_all))
```

```r
#Building a logistic regression model on training data after eliminating unimportant variables
model2 <- glm(CLASS ~ HbA1c + Chol + TG + BMI, family = binomial, data = train)
summary(model2)
```

```r
par(mfrow=c(2,2))
plot(model2)
```

```r
#checking the vif value
print(vif(model2))
```

```r
#Fitting the new model with transformed variables
model3 <- glm(CLASS ~ HbA1c + log(Chol) + TG + sqrt(BMI), family = binomial, data = train)
summary(model3)
```

```r
par(mfrow=c(2,2))
plot(model3)
```

```r
#Installing required packages
install.packages("caret", dependencies = TRUE)
install.packages('tibble')
library(caret)
```

```r
#Making predictions on the testing set for our final model
pred <- predict(model3, newdata = test, type = "response")
#Converting the probabilities to predicted classes
pred_class <- ifelse(pred > 0.5, 1, 0)
#Creating a confusion matrix
#table(test$CLASS, pred_class)
conf_matrix <- table(test$CLASS, pred_class)
conf_matrix
```

```r
#Accuracy
accuracy <- sum(diag(conf_matrix))/sum(conf_matrix)
accuracy
```

```r
#Precision
precision <- conf_matrix[2,2]/sum(conf_matrix[,2])
precision
```

```r
#Recall
recall <- conf_matrix[2,2]/sum(conf_matrix[2,])
recall
```

```r
#F1 Score
f1 <- 2 * precision * recall / (precision + recall)
f1
```

```r
#AUC-ROC
library(caTools)
library(pROC)
roc <- roc(test$CLASS, pred)
auc <- auc(roc)
auc
```

```r
#Testing the final model on the 'Diabetes==P' cases
predicted_values_p <- predict(model3, newdata = diabetes_data_P, type = "response")
prob_diabetic <- mean(predicted_values_p > 0.5)
cat("Probability of diagnosing the 'P' class cases as diabetic: ", prob_diabetic, "\n")
```

```r
#Checking if the assumptions are met or not
#Loading necessary libraries
library(car)
library(ggplot2)
library(dplyr)
```

```r
#Checking for multicollinearity
vif_values <- vif(model3)
vif_data <- data.frame(variable = names(vif_values), vif = vif_values)
print(vif_data)
```

```r
#Checking for linearity
ggplot(data = train, aes(x = AGE, y = CLASS)) + 
  geom_point() + 
  geom_smooth(method = "glm", family = "binomial") + 
  ggtitle("Linearity check: Age vs. Diabetes") + 
  xlab("Age") + 
  ylab("Probability of Diabetes")
```

```r
#Checking for influential points
infl <- influenceIndexPlot(model3, id.n = 10)

# Store studentized residuals and cook's distance in a data frame
if (length(infl$infmat) > 0) {
  infl_data <- data.frame(observation = 1:nrow(train), studentized_residuals = infl$infmat, cook_distance = infl$cinfmat)
  print(infl_data)
} else {
  message("No influential points found.")
}
```

```r
#Checking for normality of residuals
residuals <- residuals(model3, type = "pearson")
qqnorm(residuals)
qqline(residuals)
shapiro.test(residuals)
```
